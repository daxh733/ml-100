```python
import numpy as np
from numpy.linalg import norm, eig, svd

# 1. Vector Representation
v1 = np.array([2, 3])
v2 = np.array([4, -1])
print("Vector v1:", v1)
print("Vector v2:", v2)

# 2. Dot Product
dot_product = np.dot(v1, v2)
print("Dot Product of v1 and v2:", dot_product)

# 3. Matrix Multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 3]])
matrix_product = np.dot(A, B)
print("Matrix Product A @ B:\n", matrix_product)

# 4. Vector Norm (Magnitude)
norm_v1 = norm(v1)
print("Norm of v1:", norm_v1)

# 5. Orthogonality Check (dot product == 0)
are_orthogonal = np.dot(v1, v2) == 0
print("Are v1 and v2 orthogonal?", are_orthogonal)

# 6. Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = eig(A)
print("Eigenvalues of A:", eigenvalues)
print("Eigenvectors of A:\n", eigenvectors)

# 7. Singular Value Decomposition (SVD)
U, S, VT = svd(A)
print("SVD U:\n", U)
print("SVD Singular values:", S)
print("SVD V^T:\n", VT)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import norm, eig, svd

# ----- Vectors -----
v1 = np.array([2, 3])
v2 = np.array([4, -1])

# Plot vector representation
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='v2')
plt.xlim(-1, 5)
plt.ylim(-2, 4)
plt.grid()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title("Vector Representation")
plt.show()

# ----- Dot Product -----
dot_product = np.dot(v1, v2)
print("Dot Product of v1 and v2:", dot_product)

# Show cosine similarity using angle
cos_theta = dot_product / (norm(v1) * norm(v2))
theta = np.arccos(np.clip(cos_theta, -1.0, 1.0))
print(f"Angle between v1 and v2 (in degrees): {np.degrees(theta):.2f}")

# ----- Matrix and Eigenvalues -----
A = np.array([[2, 1], [1, 2]])
eigenvalues, eigenvectors = eig(A)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)

# Plot eigenvectors
origin = [0, 0]
plt.figure(figsize=(6, 6))
plt.quiver(*origin, *eigenvectors[:, 0], color='green', scale=1, scale_units='xy', angles='xy', label='Eigenvector 1')
plt.quiver(*origin, *eigenvectors[:, 1], color='orange', scale=1, scale_units='xy', angles='xy', label='Eigenvector 2')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.grid()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title("Eigenvectors of Matrix A")
plt.savefig("plots.png")
plt.show()
```
